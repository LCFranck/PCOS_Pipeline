```python
from pyspark.sql import functions as F

# Load the Bronze data
df_bronze = spark.table("bronze.pcos_data")

# Perform transformations
df_silver = df_bronze.select(
    F.col("age").cast("int"),
    F.col("bmi").cast("float"),
    
    # Standardizing Menstrual Irregularity to Boolean
    F.col("menstrual_irregularity").cast("boolean").alias("is_menstrual_irregular"),
    
    # Cleaning numeric values
    F.col("testosterone_level_ng_dl").cast("float").alias("testosterone_level"),
    F.col("antral_follicle_count").cast("int").alias("follicle_count"),
    
    # Standardizing PCOS Diagnosis to Boolean
    F.col("pcos_diagnosis").cast("boolean").alias("has_pcos")
)

# 3. Save to the Silver Layer
df_silver.write.format("delta").mode("overwrite").saveAsTable("silver.pcos_cleaned")

print("Silver table 'silver.pcos_cleaned' has been successfully updated!")
```
