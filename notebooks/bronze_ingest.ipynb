```python
import re
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

# Read the raw data
raw_df = spark.table("pcos_dataset")

# Sanitize column names: 
# Replace spaces, dots, slashes, and parentheses with underscores, and convert to lowercase
new_columns = [re.sub(r'[ ,;{}()\n\t=/]+', '_', c).strip('_').lower() for c in raw_df.columns]

# Apply the new names to the DataFrame
bronze_df = raw_df.toDF(*new_columns)

# Save to Bronze (overwrite schema can be included if something went wrong and yoiu need to restart)
bronze_df.write.format("delta").mode("overwrite").saveAsTable("bronze.pcos_data")

# Display table
display(spark.table("bronze.pcos_data").limit(5))
```
